Imputing  missing value with mean :

```python
df = pd.read_csv('titanic.csv')

print("\nMissing Values Count in Each Column:")
print(df.isnull().sum())

df_mean = df

df_mean['Age'] = df_mean['Age'].fillna(df_mean['Age'].mean())
# df_mean['Cabin'] = df_mean['Cabin'].fillna(df_mean['Unknown'].mean())

print("\nDataFrame after imputing missing values with mean:")
print(df_mean.head())
```

Impute missing values with the mode :

```python
df_mode = df
df_mode['Age'].fillna(df_mode['Age'].mode()[0], inplace=True)


print("\nDataFrame after imputing categorical columns with mode:")
print(df_mode)
```

### 2. Handle categorical value


```python
import pandas as pd


df = pd.read_csv('titanic.csv')



df['Sex'] = df['Sex'].fillna(df['Sex'].mode()[0]) 
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])  

print("\nDataFrame after imputing missing categorical columns with mode:")
print(df.head())
```

Labeling  Encoding  vs ordinal Encoding :


```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Age'] = label_encoder.fit_transform(df['Age'])

print("\nDataFrame after Label Encoding for 'Age':\n", df)
```

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

df = pd.read_csv('titanic.csv')

preprocessor = ColumnTransformer(
    transformers=[('onehot', OneHotEncoder(), ['Sex', 'Embarked'])],
    remainder='passthrough'  
)

transformed_data = preprocessor.fit_transform(df)

feature_names = preprocessor.get_feature_names_out()
df_encoded = pd.DataFrame(transformed_data, columns=feature_names)

print("\nDataFrame after One-Hot Encoding:")
print(df_encoded.head())
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, MaxAbsScaler

np.random.seed(42)
data = {
    'Age': np.random.randint(18, 80, 10),           
    'Fare': np.random.randn(10) * 30 + 50,        
    'Pclass': np.random.choice([1, 2, 3], 10),   
    'SibSp': np.random.randint(0, 5, 10),          
    'Parch': np.random.randint(0, 5, 10),   
    'Sex': np.random.choice(['male', 'female'], 10), 
    'Embarked': np.random.choice(['C', 'Q', 'S'], 10) 
}

df = pd.DataFrame(data)

standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
robust_scaler = RobustScaler()
maxabs_scaler = MaxAbsScaler()

numerical_cols = ['Age', 'Fare', 'Pclass', 'SibSp', 'Parch']

df_standard = df
df_standard[numerical_cols] = standard_scaler.fit_transform(df[numerical_cols])

df_minmax = df
df_minmax[numerical_cols] = minmax_scaler.fit_transform(df[numerical_cols])

df_robust = df
df_robust[numerical_cols] = robust_scaler.fit_transform(df[numerical_cols])

df_maxabs = df
df_maxabs[numerical_cols] = maxabs_scaler.fit_transform(df[numerical_cols])

df_log = df
df_log[numerical_cols] = np.log1p(df[numerical_cols])

# Display results
print("\nStandard Scaled Data:\n", df_standard)
print("\nMinMax Scaled Data:\n", df_minmax)
print("\nRobust Scaled Data:\n", df_robust)
print("\nMaxAbs Scaled Data:\n", df_maxabs)
print("\nLog Transformed Data:\n", df_log)
```

### 4. remove outliers


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")

numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']

# Generate some random data similar to Titanic for illustration (in case you want to test this independently)
# np.random.seed(42)
# data = np.random.normal(loc=500, scale=100, size=100)  # Normal distribution
# outliers = np.array([10, 120, 130, 5, 140, 3])  # Extreme outliers (including very low values)
# data_with_outliers = np.concatenate([data, outliers])
# df = pd.DataFrame({'Value': data_with_outliers})

# Calculate Q1, Q3, and IQR, then remove outliers for each numerical column
fig, axes = plt.subplots(1, len(numerical_cols), figsize=(15, 5))

for i, col in enumerate(numerical_cols):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_filtered = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

    ax = axes[i]
    ax.boxplot([df[col], df_filtered[col]], labels=['Original', 'Filtered'])
    ax.set_title(f"Outlier Removal: {col}")

plt.show()

print("Data after removing outliers:")
print(df_filtered[numerical_cols].head())
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")

data_reshaped = df[['Age']].dropna()

scaler = StandardScaler()
z_scores = scaler.fit_transform(data_reshaped)

z_scores = z_scores.flatten()

threshold = 3

df_filtered_z = df.loc[data_reshaped.index[np.abs(z_scores) < threshold]]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.boxplot(df['Age'].dropna())  
ax1.set_title("Before Removing Outliers")

ax2.boxplot(df_filtered_z['Age'].dropna())  
ax2.set_title("After Removing Outliers")

plt.show()

print("Data after removing outliers (Z-score method):")
print(df_filtered_z[['Age']].head())
```

### 5. feature selection 


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("titanic.csv")

numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

corr_matrix = df[numerical_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.show()

corr_threshold = 0.9
to_drop = [column for column in corr_matrix.columns if any(abs(corr_matrix[column]) > corr_threshold) and column != column]
print(to_drop)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

df = pd.read_csv('titanic.csv')

features = ['Age', 'Fare', 'Pclass', 'SibSp', 'Parch']  
df_selected = df[features].dropna()  

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_selected)

pca = PCA(n_components=5)
pca_features = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(pca_features, columns=[f'PC{i+1}' for i in range(5)])

explained_variance = pca.explained_variance_ratio_

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), explained_variance.cumsum(), marker='o', linestyle='--')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance by PCA Components")
plt.show()

print("Explained variance ratio for each component:", explained_variance)
```

```python
optimal_components = np.argmax(explained_variance >= 0.95) + 1
print(f"Optimal number of components: {optimal_components}")
pca = PCA(n_components=optimal_components)
pca
```
