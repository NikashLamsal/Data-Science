### Selecting Rows by Index

* Use .loc[] (label-based) and .iloc[] (position-based):

````python
df.loc[2]                #  (by label) by index no 
````

````python
df.iloc[2]               # (by position) always number
````

````python
df
````

````python
df.loc[8,"Film"]
````

````python
df.iloc[8,1]
````

````python
df.loc[0:3,["Actor","Film"]]
````

````python
df.iloc[0:3,0:3]
````

````python
df.iloc[0:3,[0,1,2,3,4]]
````

### Fast Access: .at and .iat


````python
df.at[0,"Actor"] ## Fast label-based access
````

````python
df.iat[0,1] ## Fast position-based access
````

### Filtering with Conditions


````python
df[df["IMDb"]>5]
````

````python
df[(df["IMDb"] > 7) & (df["Genre"] == "Action")]
````

````python
df[(df["Genre"] == "Action") | (df["Genre"] == "Comedy")]
````

#### Querying with .query()


````python
df.query("IMDb > 6 and Genre == 'Action'")
````

*  Use backticks for column names with spaces or special characters
     
```
     df.query("`first name` == 'Alice'")
```

*  You can use @ to reference Python variables
   
```
age_limit = 30
df.query("age > @age_limit") 
```
