## Creating DataFrames


### From Python Lists


```python
import pandas as pd
```

```python
data = [
    ["Alice" , 24],
    ["bob",39],
    ["charlie",34]
]
```

```python
df = pd.DataFrame(data,columns=["name","age"])
```

```python
df
```

### From Dictionary of Lists


```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35]
}

df = pd.DataFrame(data)
```

```python
df
```

## EDA (Exploratory Data Analysis)

Exploratory Data Analysis (EDA) is an essential first step in any data science project.

It involves taking a deep look at the dataset to understand its structure, spot patterns, identify anomalies, and uncover relationships between variables. This process includes generating summary statistics, checking for missing or duplicate data, and creating visualizations like histograms, box plots, and scatter plots. 

- The goal of EDA is to get a clear picture of what the data is telling you before applying any analysis or machine learning models.
