## Creating DataFrames


### From Python Lists


```python
import pandas as pd
```

```python
data = [
    ["Alice" , 24],
    ["bob",39],
    ["charlie",34]
]
```

```python
df = pd.DataFrame(data,columns=["name","age"])
```

```python
df
```

### From Dictionary of Lists


```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35]
}

df = pd.DataFrame(data)
```

```python
df
```

### Each key becomes a column, and each list is the column data.
