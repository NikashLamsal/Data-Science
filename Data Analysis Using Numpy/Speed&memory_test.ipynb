#### Using list 

####    Example 1

```python
import numpy as np
import time 

size = 1_000_000

l1 = list(range(size))
l2 = list(range(size))

start = time.time()
# list comprehension
add = [x+y for x,y in zip(l1,l2)] # Create a list of x + y for every x, y pair from zip(l1, l2)
end = time.time()
print(add[0:10])
print(end - start)
```

#### Example 2

#### Using Numpy Arrays

#### Example 1

```python
import numpy as np
import time 

size = 1_000_000

l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()
add = l1 + l2
end = time.time()
print(add[0:10])

print(end - start)
```

### Example 2

```python
start = time.time()
size = 1_000_000

l1 = np.array(list(range(size)))

numpy_squares = l1**2
end = time.time()
print(end - start)
```
