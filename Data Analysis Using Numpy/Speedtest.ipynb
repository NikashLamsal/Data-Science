#### Using list 

```python
import numpy as np
import time 

size = 1_000_000

l1 = list(range(size))
l2 = list(range(size))

start = time.time()
# list comprehension
add = [x+y for x,y in zip(l1,l2)] # Create a list of x + y for every x, y pair from zip(l1, l2)
end = time.time()
print(add[0:10])
print(end - start)
```

#### Using Numpy Arrays

```python
import numpy as np
import time 

size = 1_000_000

l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()
# list comprehension
add = l1 + l2 # Create a list of x + y for every x, y pair from zip(l1, l2)
end = time.time()
print(add[0:10])

print(end - start)
```
