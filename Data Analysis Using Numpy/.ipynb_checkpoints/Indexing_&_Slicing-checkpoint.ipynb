### Indexing and slicing


```python
import numpy as np
```

```python
arr = np.array([[1,2,3],[4,5,6],[6,7,8],[9,10,11]])
```

```python
flat = arr.flatten()
flat
```

#### Indexing (Same as Python Lists)


```python
flat[0]
```

```python
flat[-2]
```

#### Slicing (Extracting Parts of an Array)


```python
flat[2:4]
```

```python
flat[:5] # same as flat[0:5]
```

```python
flat[5:] # flat[5:last]
```

```python
flat[::2] # every second element 
```

##### Slicing returns a view, not a copy! Changes affect the original array.**

#####  In NumPy, slicing returns a view of the original array. Both the sliced array and the original array share the same data in memory, so changes in the slice affect the original array.

#### It's demo

```python
flat
```

```python
b = flat[0:3]
b
```

```python
b[1] = 98
```

```python
b
```

```python
flat
```

#### for copy

```python
b = flat[1:4].copy() # this will make copy and slice wont work 
```
